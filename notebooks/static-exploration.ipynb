```r
ransomware_df %>% 
  select(Machine, NumberOfSections, PointerToSymbolTable, NumberOfSymbols, SizeOfOptionalHeader, Characteristics, Magic,
         MajorLinkerVersion, MinorLinkerVersion, SizeOfCode, SizeOfInitializedData, SizeOfUninitializedData, AddressOfEntryPoint,
         BaseOfCode, BaseOfData, ImageBase, SectionAlignment, FileAlignment, MajorOperatingSystemVersion, MinorOperatingSystemVersion,
         MajorImageVersion, MinorImageVersion, MajorSubsystemVersion, MinorSubsystemVersion, Reserved1, SizeOfImage, SizeOfHeaders,
         CheckSum, Subsystem, DllCharacteristics, SizeOfStackReserve, SizeOfStackCommit, SizeOfHeapReserve, SizeOfHeapCommit, LoaderFlags, NumberOfRvaAndSizes) %>%
  mutate(Dummy = 1) %>%
  pivot_longer(everything(), names_to = "Feature", values_to = "Value") %>%
  group_by(Feature) %>%
  summarize(Min = min(Value),
            Max = max(Value),
            Mean = mean(Value),
            Median = median(Value),
            Q1 = quantile(Value, .25),
            Q3 = quantile(Value, .75),
            P40 = quantile(Value, .40),
            Range = max(Value) - min(Value),
            Var = var(Value),
            Std = sd(Value),
            DMA = mean(abs(Value - mean(Value))),
            DMedA = median(abs(Value - mean(Value))),
            IQ = quantile(Value, 0.75) - quantile(Value, 0.25))
```

# Bar Charts

```r
sransomware_df <- 
  ransomware_df %>%
  select(Machine, NumberOfSections, PointerToSymbolTable, NumberOfSymbols, SizeOfOptionalHeader, Characteristics,
         Magic, MajorLinkerVersion, MinorLinkerVersion, SizeOfCode, SizeOfInitializedData, SizeOfUninitializedData) %>%
  pivot_longer(everything(), names_to = "Feature", values_to = "Value") %>%
  group_by(Feature, Value) %>%
  summarize(Freq = n())

ggplot(sransomware_df, aes(x = Value, y = Freq, fill = Feature)) +
  geom_bar(stat = "identity") +
  facet_wrap(vars(Feature), scales = "free")
```

```r
sransomware_df <- 
  ransomware_df %>%
  select(AddressOfEntryPoint,BaseOfCode, BaseOfData, ImageBase, SectionAlignment, FileAlignment, MajorOperatingSystemVersion,
         MinorOperatingSystemVersion, MajorSubsystemVersion, MinorSubsystemVersion, Reserved1, SizeOfImage) %>%
  pivot_longer(everything(), names_to = "Feature", values_to = "Value") %>%
  group_by(Feature, Value) %>%
  summarize(Freq = n())

ggplot(sransomware_df, aes(x = Value, y = Freq, fill = Feature)) +
  geom_bar(stat = "identity") +
  facet_wrap(vars(Feature), scales = "free")
```

```r
sransomware_df <- 
  ransomware_df %>%
  select(SizeOfHeaders, CheckSum, Subsystem, DllCharacteristics, SizeOfStackReserve, SizeOfStackCommit,
         SizeOfHeapReserve, SizeOfHeapCommit, LoaderFlags, NumberOfRvaAndSizes) %>%
  pivot_longer(everything(), names_to = "Feature", values_to = "Value") %>%
  group_by(Feature, Value) %>%
  summarize(Freq = n())

ggplot(sransomware_df, aes(x = Value, y = Freq, fill = Feature)) +
  geom_bar(stat = "identity") +
  facet_wrap(vars(Feature), scales = "free")
```
